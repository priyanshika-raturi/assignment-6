2. How does broadcasting work in NumPy?

**Broadcasting** in NumPy is a rule that allows operations on arrays of **different shapes**. Instead of throwing an error when shapes don’t match, NumPy **automatically expands** (broadcasts) the smaller array so it can work element-wise with the larger one.

How it works:

1. Compare the shapes of the two arrays from right to left.
2. Dimensions are compatible if:

   * They are **equal**, or
   * One of them is **1**
3. If a dimension is **1**, NumPy stretches it to match the other dimension.

Example:


 Adding arrays of different shapes

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

print(a + b)
```

**How broadcasting works:**

* `a` has shape (2, 3)
* `b` has shape (3) → treated as (1, 3)
* NumPy expands `b` to (2, 3) and then adds element-wise.

4.Explain the use of the groupby() method in Pandas?


The groupby() method in Pandas is used to split data into groups based on one or more columns and then perform operations such as sum, mean, count, or any custom function on each group.


How groupby() works:

It follows the **split → apply → combine** process:

1. Split the data into groups based on a column.
2. Apply a function (like sum, mean, count).
3. Combine the results back into a new DataFrame.



Example:

```python
import pandas as pd

data = {
    'Category': ['A', 'A', 'B', 'B'],
    'Values': [10, 20, 30, 40]
}

df = pd.DataFrame(data)

result = df.groupby('Category')['Values'].sum()
print(result)
```

Output:

```
Category
A    30
B    70
```

6.What are the differences between NumPy arrays and Python lists?

1. Performance

* **NumPy arrays** are much **faster** because they use optimized C code.
* **Python lists** are slower because they store elements as objects and require more memory.

2. Memory Efficiency

* **NumPy arrays** use **less memory** and store data in a continuous block.
* **Python lists** use **more memory** because each element is a separate Python object.

3. Type of Elements

* **NumPy arrays:** All elements must be of the **same data type** (e.g., all integers or all floats).
* **Python lists:** Can contain **different data types** (e.g., int, string, float).

4. Mathematical Operations

* **NumPy arrays:** Support **vectorized operations** (e.g., `a + b` adds element-wise).
* **Python lists:** Do **not** support element-wise math without loops.

5. Functionality

* **NumPy arrays:** Have many built-in functions for math, statistics, linear algebra, reshaping, etc.
* **Python lists:** Provide basic operations only (append, remove, slice).

6. Multi-Dimensional Support**

* **NumPy arrays:** Support **multi-dimensional arrays** (2D, 3D, …).
* **Python lists:** Need nested lists for multi-dimensional data (messy and inefficient).

8.What does the term “vectorized operation” mean in NumPy?


A vectorized operation in NumPy means performing operations on entire arrays at once, without using explicit Python loops. NumPy applies the operation to each element internally using fast, optimized C code.

Why it is useful:

* Much faster than using loops in Python
* Cleaner and shorter code
* Allows efficient mathematical operations on large datasets

Example:

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)   # vectorized addition
```

Here, `a + b` adds all elements without writing a loop.

21.Why is NumPy’s array slicing faster than Python’s list slicing?

NumPy’s array slicing is faster mainly because of how NumPy stores and processes data. The key reasons are:


1. NumPy Uses Contiguous Memory Blocks

NumPy arrays store data in a single continuous block of memory, while Python lists store elements in separate memory locations.
Because of this:

* NumPy can access elements much faster
* Slicing becomes a simple pointer adjustment rather than copying data

2. Slicing Returns a “View,” Not a Copy

When you slice a NumPy array, it does not copy the data.
It creates a view that points to the same memory.

But Python list slicing creates a new list, which takes more time.

3. NumPy Is Written in Optimized C Code

NumPy operations (including slicing) run in fast, low-level C code, whereas Python lists operate through slower Python-level loops.


4. Fixed Data Types

All elements in a NumPy array have the same data type, so accessing and slicing are simpler and faster.
Python lists store objects of different types, adding overhead.
